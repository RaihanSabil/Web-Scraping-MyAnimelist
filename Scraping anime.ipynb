```python
import requests
from bs4 import BeautifulSoup
import time
from random import randint
import pandas as pd

############## UNIVERSAL FUNCTIONS ##############

def get_soup(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    return soup

############## END OF UNIVERSAL FUNCTIONS ##############

df_anime = {
  'Name': [],
  'Favorite': [],
  'Rating' : [],
  'AniName': [],
  'FavChar': [],
  'CharName': [],
  'Ranked': [],
  'Popularity': [],
  'Members': []

}

source = 'https://myanimelist.net/people.php?limit=0'
soup = get_soup(source)

list_name = []
list_fav = []
list_rate = []
list_fav2 = []
list_AniName = []
list_FavChar = []
list_rating = []
list_CharName = []
list_Ranked = []
list_Popularity = []
list_Members = []

Name = 0
fav = 0
Normalized = 0

seiyuus = soup.find_all('td', {'class': 'people'}) #scrape tag td dengan class 'people'
for a in seiyuus:
    time.sleep(10)
    PeopleHref = a.findChild('a', {'class':'fs14 fw-b'})#scrape tag a dengan class 'fs14 fw-b'
    if PeopleHref is None:
        continue
    print(PeopleHref) #menunjukkan hasil SeiyuuHref
    LinkOfPeople = PeopleHref.get('href')# mengambil tag href dari SeiyuuHref
    print(LinkOfPeople)
    soup2 = get_soup(LinkOfPeople)
    arr_TR = soup2.find_all('tr', {'class': 'js-people-character js-anime-watch-status-people-va-notinmylist'})#menjadikan class ini sebagai parameter seiyuu

    for a in arr_TR: #mengambil data nama karakter anime
        char = a.find_all('td', {'class':'borderClass', 'align':'right'})[0]
        if char is None:
            continue
        charname = char.findChild('a').getText()
        print(charname)
        list_CharName.append(charname)#memasukkan data nama karakter anime kedalam list

    for a in arr_TR: #mengambil data jumlah favorite karakter anime
        Favchar = a.findChild('small', {'class':'js-people-favorites'})
        if Favchar is None:
            continue
        else:
            FavChar2 = int(Favchar.getText())
            list_FavChar.append(FavChar2)#memasukkan data jumlah favorite karakter anime kedalam list
            print(FavChar2)

    for b in arr_TR: #mengambil data judul anime
        ListOfAnimes = b.findChild('a', {'class':'js-people-title'})
        if ListOfAnimes is None:
            continue
        else:
            arr_fav = soup2.find_all('td', {'class':'borderClass'})
            for i in arr_fav:#mengambil data jumlah favorite seiyuu
                Fav = i.find_all('div', {'class':'spaceit_pad'})
                Fav = Fav[-1].getText().split(' ')[-1]
                a= Fav.replace(",","")
                list_fav.append(a)
                newfav =  [int(i) for i in list_fav]
                fav = newfav[-1]
                break
            arr_name = soup2.find_all('div', {'class':'h1 edit-info'})
            for i in arr_name: #mengambil data nama seiyuu
                Name = i.findChild('h1', {'class':'title-name h1_bold_none'}).getText()
            AnimeName = ListOfAnimes.getText() #mengambil data judul anime
            LinkOfAnimes = ListOfAnimes.get('href')#mengambil data tautan halaman anime

            soup3 = get_soup(LinkOfAnimes)
            arr_c = soup3.find_all('div', {'class':['score-label score-na','score-label score-1','score-label score-2','score-label score-3','score-label score-4','score-label score-5', 'score-label score-6', 'score-label score-7',
                                                    'score-label score-8', 'score-label score-9', 'score-label score-10']})#mengambil data rating
            for c in arr_c:
                time.sleep(2)
                rate = c.getText()
                if rate == "N/A":# memrubah nilai "N/A" menjadi 0
                    rate = 0
                print(AnimeName)
                list_AniName.append(AnimeName) #memasukkan data judul anime
                Normalized = float(rate)#mengubah format string menjadi float
                print(Normalized)
                list_rating.append(Normalized)#memasukkan data rating anime berformat float
                Name =  Name
                print(Name)
                list_name.append(Name)#memasukkan data nama seiyuu kedalam list
                fav = fav
                print(fav)
                list_fav2.append(fav)#memasukkan data jumlah favorite seiyuu kedalam list

            Pop = soup3.find_all('div', {'class':'di-ib ml12 pl20 pt8'})
            for i in Pop:#mengambil data rank, popularity, dan jumlah member
                Ranked = i.find_all('strong')[0].getText()
                if Ranked == "N/A":
                    Ranked = 0
                else:
                    strip = Ranked.replace("#","")
                    Ranked = int(strip)
                list_Ranked.append(Ranked)#memasukkan data rank kedalam list
                print(Ranked)
                Popularity = i.find_all('strong')[1].getText()
                strip = Popularity.replace("#","")
                Popularity = int(strip)
                list_Popularity.append(Popularity)#memasukkan data popularity kedalam list
                print(Popularity)
                Members = i.find_all('strong')[2].getText()
                strip = Members.replace(",","")
                Members = int(strip)
                list_Members.append(Members)#memasukkan data jumlah member kedalam list
                print(Members)

print(len(list_name))
print(len(list_fav2))
print(len(list_rating))
print(len(list_CharName))
print(len(list_FavChar))
print(len(list_Ranked))
print(len(list_Popularity))
print(len(list_Members))

df_anime['Name'] = list_name
df_anime['Favorite'] = list_fav2
df_anime['Rating'] =  list_rating
df_anime['AniName'] = list_AniName
df_anime['CharName'] = list_CharName
df_anime['FavChar'] = list_FavChar
df_anime['Ranked'] = list_Ranked
df_anime['Popularity'] = list_Popularity
df_anime['Members'] = list_Members

df_anime = pd.DataFrame(df_anime)
df_anime.to_csv('AnimeList3.csv', index=False)
```
